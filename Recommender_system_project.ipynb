```python
import ast

# Helper function to extract names from json in list form
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L   
```

- Total 5000 vectors are created and each vector is representing single movie, now we will calculate distance between each movie to show similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_metrics = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity_metrics[1])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies[movies["title"]==movie].index[0]
    distance = similarity_metrics[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
    
```
